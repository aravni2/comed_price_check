```python
def convert_time(str_time):
    time_ms = int(str_time)
    time_cst = dt.datetime.fromtimestamp(time_ms/1000.0).strftime('%Y-%m-%d %I:%M')

    return time_cst

def get_price (interval = 0):
    # retrieve response
    resp_5min = requests.get('https://hourlypricing.comed.com/api?type=5minutefeed')
    resp_5min.json()
    
    # convert time
    time_5m = int(resp_5min.json()[interval]['millisUTC']) 
    time_cst = dt.datetime.fromtimestamp(time_5m/1000.0).strftime('%Y-%m-%d %I:%M')
    
    
    cur_price_5m = resp_5min.json()[interval]['price']
    price_diff = float(resp_5min.json()[interval]['price']) - float(resp_5min.json()[interval+1]['price'])

    return (time_cst,cur_price_5m,price_diff)


# resp = requests.get('https://hourlypricing.comed.com/api?type=currenthouraverage')
# resp_5min = requests.get('https://hourlypricing.comed.com/api?type=5minutefeed')
# resp_5min.json()


# time_5m = resp_5min.json()[0]['millisUTC']
# cur_price_5m = resp_5min.json()[0]['price']
# price_diff = float(resp_5min.json()[0]['price']) - float(resp_5min.json()[1]['price'])
# time_cst_5m = convert_time(time_5m)

# time = resp.json()[0]['millisUTC']
# time_cst = convert_time(time)
# hr_price = resp.json()[0]['price']

# print(dt.datetime.utcfromtimestamp(time_ms / 1000.),dt.UTC)

prev_time, prev_price_5m, prev_price_diff = get_price(1)
time_cst_5m, cur_price_5m, price_diff = get_price()

print(f'Current Interval: {time_cst_5m}, current price: {cur_price_5m}, diff: {price_diff}')
print(f'previous Interval: {prev_time}, previous price: {prev_price_5m}, previous diff: {prev_price_diff}')
# print(f'Hour Avg: {time_cst}, hourly price: {hr_price}')
```

```python
def get_price (interval = 0):
    # retrieve response
    resp_5min = requests.get('https://hourlypricing.comed.com/api?type=5minutefeed')
    resp_5min.json()
    
    # convert time
    time_5m = int(resp_5min.json()[interval]['millisUTC']) 
    time_cst = dt.datetime.fromtimestamp(time_5m/1000.0).strftime('%Y-%m-%d %I:%M')
    
    
    cur_price_5m = resp_5min.json()[interval]['price']
    price_diff = float(resp_5min.json()[interval]['price']) - float(resp_5min.json()[interval+1]['price'])

    return (time_cst,cur_price_5m,price_diff)



# initiate pygame
pygame.init()
channel = pygame.mixer.Channel(0)

s_ding = pygame.mixer.Sound('airplane_ding.wav')
d_ding = pygame.mixer.Sound('airplane_ding_ding.wav')

while True:
    try:
        prev_time, prev_price_5m, prev_price_diff = get_price(1)
        time_cst_5m, cur_price_5m, price_diff = get_price()

        print(f'Current Interval: {time_cst_5m}, current price: {cur_price_5m}, diff: {price_diff}')
        print(f'previous Interval: {prev_time}, previous price: {prev_price_5m}, previous diff: {prev_price_diff}')
        if(float(cur_price_5m) <5):
            channel.play(s_ding,0)
    except Exception as e:
        print(f"Request failed: {e}")
    
    time.sleep(60)  # Wait for 60 seconds
```
